```python
# Dependencies
import pandas as pd
import numpy as np
import requests
import json
import requests
import urllib.request
from pandas.io.json import json_normalize
```

```python
# Import coins file as DataFrame
coins_df = pd.read_csv("../resources/coingecko_symbols.csv")
```

```python
# Add columns for sentiment
# Note that we used "" to specify initial entry.
coins_df["avg_sentiment"] = ""
coins_df["avg_sentiment_1y"] = ""
coins_df["avg_sentiment_pct_chg"] = ""
coins_df["social_score_1y"] = ""
coins_df["social_score_1y_pct_chg"] = ""
coins_df
```

```python
base_url = "https://api.lunarcrush.com/v2?data=assets&key=b9c27jge0ku5kvggiqykth&change=1y&data_points=0"
```

```python
# Use the coins_df file to select symbol from sentiment data
for index, row in coins_df.iterrows():
    # get coin symbol
    symbol = row["symbol"]

    # Use the search term: "Coin" and our symbol
    url = f"{base_url}&symbol={symbol}"

    # make request and print url
    response = requests.get(url).json()
    

#     # Since some data may be missing we incorporate a try-except to skip any that are missing a data point.
    try:
        coins_df.loc[index,"avg_sentiment"] = response["data"][0]["average_sentiment"]
        coins_df.loc[index,"avg_sentiment_1y"] = response["data"][0]["average_sentiment_1y"]
        coins_df.loc[index,"avg_sentiment_pct_chg"] = response["data"][0]["average_sentiment_1y_percent_change"]
        coins_df.loc[index,"social_score_1y"] = response["data"][0]["social_score_1y"]
        coins_df.loc[index,"social_score_1y_pct_chg"] = response["data"][0]["social_score_1y_percent_change"]
        
    except (KeyError, IndexError):
        print("Missing field/result... skipping.")
```
