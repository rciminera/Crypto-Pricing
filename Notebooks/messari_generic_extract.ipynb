```python
# Dependencies
import pandas as pd
import numpy as np
import requests
import json
import requests
import urllib.request
from pandas.io.json import json_normalize
```

```python
# Import coins file as DataFrame
coins_df = pd.read_csv("../resources/coingecko_symbols.csv")
```

```python
dict = {'symbol': 'Symbol',
        'id': 'ID',}
coins_df.rename(columns=dict,
          inplace=True)
coins_df
```

```python
# Add columns for sentiment
# Note that we used "" to specify initial entry.
coins_df["Category"] = ""
coins_df["Sector"] = ""
coins_df["Token Type"] = ""
coins_df["Token Usage"] = ""
coins_df["Consensus Algorithm"] = ""
coins_df["Summary"] = ""
coins_df
```

```python
coins_df = coins_df.iloc[249:251]
coins_df
```

```python
base_url = "https://data.messari.io/api/v2/assets/"
# https://data.messari.io/api/v2/assets/{assetKey}/profile
base_url
```

```python

header = {"x-messari-api-key": "291b4bba-a9be-4e80-a3d3-8131b1208c8e"}
header
```

```python
# Use the coins_df file to select symbol from sentiment data
for index, row in coins_df.iterrows():
    # get coin symbol
    symbol = row["Symbol"]

    # Use the search term: "Coin" and our symbol
    url = f"{base_url}{symbol}/profile"

    # make request and print url

    response = requests.get(url, header).json()

#     # Since some data may be missing we incorporate a try-except to skip any that are missing a data point.
    try:
        coins_df.loc[index,"Category"] = response["data"]["profile"]["general"]["overview"]["category"]
        coins_df.loc[index,"Sector"] = response["data"]["profile"]["general"]["overview"]["sector"]
        coins_df.loc[index,"Token Type"] = response["data"]["profile"]["economics"]["token"]["token_type"]
        coins_df.loc[index,"Token Usage"] = response["data"]["profile"]["economics"]["token"]["token_usage"]
        coins_df.loc[index,"Consensus Algorithm"] = response["data"]["profile"]["economics"]["consensus_and_emission"]["consensus"]["general_consensus_mechanism"]
        coins_df.loc[index,"Summary"] = response["data"]["profile"]["general"]["overview"]["project_details"]
    
    
    except (KeyError, IndexError):
        print("Missing field/result... skipping.")
```
