```python
# Dependencies
import pandas as pd
import numpy as np
import requests
import json
import requests
import urllib.request
from pandas.io.json import json_normalize
```

```python
# Import coins file as DataFrame
coins_df = pd.read_csv("../resources/coingecko_symbols.csv")
```

```python
# Add columns for sentiment
# Note that we used "" to specify initial entry.
coins_df["category_1"] = ""
coins_df["category_2"] = ""
# coins_df["hashing_algorithm"] = ""
coins_df
```

coins_df = coins_df.iloc[0:50]

```python
coins_df = coins_df.iloc[200:250]
coins_df
```

```python
base_url = "https://api.coingecko.com/api/v3/coins/"
```

```python
# Use the coins_df file to select symbol from sentiment data
for index, row in coins_df.iterrows():
    # get coin id for congecko category pull
    id = row["id"]

    # Use the search term: "Coin" and id
    url = f"{base_url}{id}"

    # make request and print url
    response = requests.get(url).json()
    

#     # Since some data may be missing we incorporate a try-except to skip any that are missing a data point.
    try:
        coins_df.loc[index,"category_1"] = response["categories"][0]
        coins_df.loc[index,"category_2"] = response["categories"][1]
#         coins_df.loc[index,"hashing_algorithm"] = response["hashing_algorithm"]
        
        
    except (KeyError, IndexError):
        print("Missing field/result... skipping.")
```
